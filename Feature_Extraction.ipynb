```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
# Assuming df is your DataFrame
# Convert 'runs' column to numeric using .loc
df.loc[:, 'runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match id' and calculate the cumulative sum of 'runs' using .loc
df.loc[:, 'current_score'] = df.groupby('match id')['runs'].cumsum()

```

```python
df
```

```python
# Assuming df is your DataFrame
# Split 'ball' column into 'over' and 'ball_no' columns using lambda functions and .apply
df['over'] = df['ball'].apply(lambda x: str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x: str(x).split(".")[1])

# Use .loc to assign the values to the new columns
df.loc[:, 'over'] = df['ball'].apply(lambda x: str(x).split(".")[0])
df.loc[:, 'ball_no'] = df['ball'].apply(lambda x: str(x).split(".")[1])
```

```python
df.loc[:, 'balls_bowled'] = (df['over'].astype('int') * 6) + df['ball_no'].astype('int')
```

```python
# Calculate 'balls_left' using .loc to assign values
df.loc[:, 'balls_left'] = 120 - df['balls_bowled']
df.loc[:, 'balls_left'] = df['balls_left'].apply(lambda x: 0 if x < 0 else x)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == '0' else 1)
# Convert 'player_dismissed' column to integer type
df['player_dismissed'] = df['player_dismissed'].astype(int)

# Calculate cumulative sum of 'player_dismissed' for each match
df['player_dismissed_cumsum'] = df.groupby('match id')['player_dismissed'].transform('cumsum')

# Calculate 'wickets_left' column
df['wickets_left'] = 10 - df['player_dismissed_cumsum']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')  # 'coerce' will replace non-numeric values with NaN

# Group by 'match id'
groups = df.groupby('match id')

# Get unique match ids
match_ids = df['match id'].unique()

# Calculate rolling sum for each group and extend the last_five list
last_five = []
window_size = 30

for match_id in match_ids:
    group_data = groups.get_group(match_id)
    rolling_sum = group_data['runs'].rolling(window=window_size, min_periods=1).sum()
    last_five.extend(rolling_sum.values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match id').sum()['runs'].reset_index().merge(df,on='match id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
X = final_df.drop(columns=['runs_x', 'wickets_left'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
X_train.dropna(inplace=True)

trf = ColumnTransformer([
    ('trf', OneHotEncoder(drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler(with_mean=False)),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
