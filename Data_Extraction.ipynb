```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm 
```

```python
filenames = []  # Create an empty list to store file names

# Iterate over the files in the 'data' directory
for file in os.listdir('data'):
    # os.path.join is used to create a full path by joining the directory 'data' and the file name
    full_path = os.path.join('data', file)
    
    # Append the full path to the list of filenames
    filenames.append(full_path)
```

```python
filenames[0:5]
```

```python
final_df = pd.DataFrame()
print(type(final_df)) 
counter=1
for file in tqdm(filenames):
    with open(file,'r') as f:
        df=pd.json_normalize(safe_load(f))
        df['match_id']=counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter+=1
        
final_df
```

```python
matches=pickle.load(open('dataset_level1.pkl','rb'))
matches
```

```python
# Initialize count and an empty DataFrame for storing delivery details
count = 1
delivery_df = pd.DataFrame()

# Iterate through each row in the 'matches' DataFrame
for index, row in matches.iterrows():
    # Check if the current count is in the specified list of indices to skip
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        # Skip the current iteration and increment count
        count += 1
        continue
    
    # Increment count
    count += 1
    
    # Initialize lists to store delivery details for the current match
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = [] 
    batting_team = []
    match_id = []
    city = []
    venue = []
    
    # Iterate through each ball in the first innings of the current match
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            # Append relevant details to the corresponding lists
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            
            # Try to append player dismissed information; if not available, append '0'
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    
    # Create a DataFrame for the current match's delivery details
    loop_df = pd.DataFrame({
            'match id': match_id,
            'teams': teams,
            'batting_team': batting_team,
            'ball': ball_of_match,
            'batsman': batsman,
            'bowler': bowler,
            'runs': runs,
            'player_dismissed': player_of_dismissed,
            'city': city, 
            'venue': venue
        })
    
    # Append the current match's delivery details to the overall DataFrame
    delivery_df = pd.concat([delivery_df, loop_df], ignore_index=True)

```

```python
delivery_df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team']=delivery_df.apply(bowl,axis=1)
```
